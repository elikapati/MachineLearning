```python
# Descriptive Statistics

# Peek at your data
from pandas import read_csv

filename = 'diabetes.csv'
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = read_csv(filename, header = 0, names = names)

print("Shape of data = ", df.shape)
peek = df.head(20)
peek
```

```python
# Data type of each attribute
print(df.dtypes)
```

```python
# Examine Statistical Summary
# Check for Null values - either discard them or replace with proper values
df.describe()
```

```python
# Check how data is distributed by class
# For example,
#      Non-diabetics are twice Diabetics
class_counts = df.groupby('Outcome').size()
print("Non-diabetics = ", class_counts[0])
print("Diabetics = ", class_counts[1])
```

```python
# Check for correlations between features
# Note: Machine learning algorithms like linear and logistic regression can suffer
#       poor performance if there are highly correlated attributes in your dataset
# correlation of 0 indicates no relation
# correlation of -1 indicates inverse relation
# correlation of +1 indicates direct relation
df.corr(method = 'pearson')
```

```python
# Check for Skewness
# Many machine learning algorithms assume a Gaussian distribution. 
# Skewed data is either shifted or squashed data in some direction
# Correct the skewness in order to improve accuracy of the models
# Values closer to 0 show less skew
# +ve or -ve values are skewed
df.skew()
```

```python
# Visualizations - Histograms:
# Appears that BMI, BP, Glucose are closer to Gaussian distribution
# Appears that Age, Diabetes Pedigree, Pregnancies are closer to exponential distribution

from matplotlib import pyplot as plt

plt.figure()
df.hist()
plt.subplots_adjust(top=1, bottom=0, left=0, right=1, hspace=0.5, wspace=0.5)
plt.show()
```

```python
# Visualizations - Density Plots:
#  Density plots are another way of getting a quick idea of the distribution of each attribute. 
#  The plots look like an abstracted histogram with a smooth curve drawn through the top of each bin,
#  much like your eye tried to do with the histograms.

df.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.subplots_adjust(top=2, bottom=0, left=0, right=2, hspace=0.25, wspace=0.25)
plt.show()
```

```python
# Visualizations - Box and Whisker Plots:
#   The whiskers give an idea of the spread of the data and dots outside of the whiskers show candidate outlier values

df.plot(kind='box', subplots=True, vert=False, layout=(9,1), sharex=False)
plt.subplots_adjust(top=2, bottom=0, left=0.5, right=2, hspace=0.5, wspace=0.2)
plt.show()
```

```python
# Visualizations - Multivariate Plots (Correlation Matrix Plot)
import numpy

correlations = df.corr()

# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)

cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

ticks = numpy.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(names)
ax.set_yticklabels(names)

plt.show()
```
