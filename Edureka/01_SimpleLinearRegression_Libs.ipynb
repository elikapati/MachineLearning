```python
# Demonstration of Simple Linear Regression algorithm using Scikit Library

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Read data
# source of data https://raw.githubusercontent.com/mubaris/potential-enigma/master/headbrain.csv
data = pd.read_csv('Data//headbrain.csv')
print('Shape of input data = ', data.shape, '(Rows and Cols)')
data.head()

# Collect X and Y
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values

print(X)
print(Y)

```

```python
# Total number of values
m = len(X)
print(m)
```

```python
# Cannot use Rank 1 matrix in scikit learn
X = X.reshape((m,1))
print(X)
```

```python
# Create model
reg = LinearRegression()

# Fit training data
reg = reg.fit(X, Y)

# predict Y
Y_pred = reg.predict(X)

# Calculate RMSE and R2 scope
mse = mean_squared_error(Y, Y_pred)
rmse = np.sqrt(mse)
r2_score = reg.score(X, Y)

print('Root mean square = ', np.sqrt(mse))
print('R Squared = ', r2_score)
```
