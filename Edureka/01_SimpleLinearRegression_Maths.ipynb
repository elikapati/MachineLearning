```python
# Demonstration of Simple Linear Regression algorithm using Mathematics
# Problem:
#    Given human brain's head size and brain weight find a relation between the two

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)

# Read data
# source of data https://raw.githubusercontent.com/mubaris/potential-enigma/master/headbrain.csv
data = pd.read_csv('Data//headbrain.csv')
print('Shape of input data = ', data.shape, '(Rows and Cols)')
data.head()
```

```python

# Collect X and Y
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
print("X:\n", X)
print("Y:\n", Y)
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
m = len(X)

# Use the formula to calculate b1 and b2
numer = 0
denom = 0

# assuming y = b0 + b1 * x as the linear equation
for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

# Print coefficients
print('b1 = ', b1, 'b0 = ', b0)
print('The equation of the line is :- y = ', b0, ' + ', b1, ' * x')


# Plot values and regression line
max_x = np.max(X) + 100
min_x = np.min(X) - 100

# Calculate line values x and y; linspace gets evenly spaced '1000' values from 'min_x' to 'max_x' 
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

# Plot line
plt.plot(x, y, color='g', label='Regression Line')

# Plot scatter points
plt.scatter(X, Y, color='y', label='Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()

ss_t = 0
ss_r = 0

for i in range(m):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r / ss_t)
print('R Squared = ', r2)
```
