```python
import numpy as np
import pandas as pd
import seaborn as sns

planets = sns.load_dataset('planets')
planets.shape
planets.head

# Aggregations in pandas
# Pandas aggregates for series
rng = np.random.RandomState(42)
ser = pd.Series(rng.rand(5))
ser

print("Aggregations on pandas series:")
print("Sum of series: ", ser.sum())
print("Mean of series: ", ser.mean())

# Pandas aggregates for data frame
df = pd.DataFrame({'A': rng.rand(5),
                   'B': rng.rand(5)})
print("Dataframe:")
print(df)

print("\nAggregations on pandas dataframe:")
print("Mean of dataframe column-wise: ")
print(df.mean())
print("Mean of dataframe row-wise:")
print(df.mean(axis=1))
```
