```python
import numpy as np
import pandas as pd
import seaborn as sns

# Demo on binning using pandas cut
raw_data = {
    "city": ["Hyderabad", "Chennai", "Bangalore", "Calcutta", "Pune", "Delhi", "Bombay", "Ahmedabad", "Hubli"],
    "score": [67, 63, 55, 70, 64, 77, 45, 66, 72]
}

# Objective is to grade the city to be one of grades viz., low, average, good, brilliant
# based on the score distributed into bins
df = pd.DataFrame(raw_data, 
                  index = pd.Index(raw_data['city'], name='city'), 
                  columns = pd.Index(['score'], name='attributes'))
df
```

```python
# define bins as 0 - 25, 25 - 50, 50 - 75, 75 - 100
bins = [0, 25, 50, 75, 100]
```

```python
# define names of bins
grades = ['low', 'average', 'good', 'brilliant']
```

```python
# Using Pandas - cut -> Create and append a new column 'grade' and fill it with grades based on score
df['grade'] = pd.cut(df['score'], bins, labels=grades)
df
```

```python
# US Birth rate example
births = pd.read_csv('../data/births.csv')
births
```

```python
births['decade'] = 10 * (births['year'] // 10)
births
```

```python
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

sns.set() # use Seaborn styles
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year')
```

```python
# Remove outliers caused by typo
quartiles = np.percentile(births['births'], [25, 50, 75])
print(quartiles)
mu = quartiles[1]
print(mu)
sig = 0.74 * (quartiles[2] - quartiles[0])
print(sig)
```

```python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births
```

```python
# set 'day' column to integer; it originally was a string due to nulls
births['day'] = births['day'].astype(int)
births
```

```python
# create a datetime index from the year, month, day
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format='%Y%m%d')
births['dayofweek'] = births.index.dayofweek
births
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
births.pivot_table('births', index='dayofweek',
                             columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
```

```python
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.head()
```

```python
births_by_date.index = [pd.datetime(2012, month, day) for (month, day) in births_by_date.index]
births_by_date.head()
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax);
```
