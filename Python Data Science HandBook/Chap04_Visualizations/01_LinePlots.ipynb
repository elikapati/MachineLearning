```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# MATLAB Style interface
# Suplots using MATLAB style interfce
plt.figure # Create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# Simple Line Plots
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
# Using python lab
plt.plot(x, np.sin(x))
```

```python
# Specifying color in different ways
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# Specifying the line styles
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')   # solid
plt.plot(x, x+5, linestyle='--')  # dashed
plt.plot(x, x+6, linestyle='-.')  # dashdot
plt.plot(x, x+7, linestyle=':')   # dotted
```

```python
# combined specification of style and color
plt.plot(x, x+0, '-g')  # solid,    green
plt.plot(x, x+1, '--c') # dash,     cyan
plt.plot(x, x+2, '-.k') # dash dot, black
plt.plot(x, x+3, ':r')  # dot,      red
```

```python
# Adjusting plot Axes limits
plt.plot(x, np.sin(x))
plt.xlim(-5, 15)
plt.ylim(-1.5, 1.5);
```

```python
# Reversing the axes
plt.plot(x, np.sin(x))
plt.xlim(15, -5)      # reversed uppper and lower limits of x axis
plt.ylim(1.5, -1.5);  # reversed uppper and lower limits of y axis
```

```python
# or all limits in single call as below
plt.plot(x, np.sin(x))
plt.axis([15, -5, 1.5, -1.5]) # [xmin, xmax, ymin, ymax]
```

```python
# you can even tightly pack the curve in the figure
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# you can even maintain equal aspect ratio
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# Labeling plots
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
# Specifying the legend
plt.plot(x, np.sin(x), '-g', label='sin(x)')  # legend of sinx
plt.plot(x, np.cos(x), ':b', label='cos(x)')  # legend of cosx
plt.axis('equal')
plt.legend();
```

```python
# Transitioning from MATLAB style functions to object oriented methods
#   plt.xlabel()  ax.set_xlabel()
#   plt.ylabel()  ax.set_ylabel()
#   plt.xlim()  ax.set_xlim()
#   plt.ylim()  ax.set_ylim()
#   plt.title()  ax.set_title()
# You may call individually as stated above or combinedly as shown below:
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
        xlabel='x', ylabel='sin(x)',
        title='A Simple Plot');
```
