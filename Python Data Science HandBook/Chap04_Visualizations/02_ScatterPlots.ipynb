```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 10, 30)
x
```

```python
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
```

```python
# Demo markers
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']: 
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
# specifying the line style, points, color
plt.plot(x, y, '-ok') # line, circle, black
```

```python
# additional keyword arguments
plt.plot(x, y, '-p', color='gray', # pentagon
         markersize=15, linewidth=4, markerfacecolor='white', markeredgecolor='gray', markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

```python
# The primary difference of plt.scatter from plt.plot is that it can be used to create
# scatter plots where the properties of each individual point (size, face color, edge color,
# etc.) can be individually controlled or mapped to data.

plt.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # to show the color scale
```
