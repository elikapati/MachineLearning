```python
import numpy as np

x = np.array([2, 4, 1, 5, 3])
print("Elements of x:")
print(x)

# sort x and copy to y; leave x in tact
y = np.sort(x)
print("Sort x and copy to y (leaving x in tact):")
print(y)
print("Unaffected x:")
print(x)

# in-place sorting of x
print("In-place sorting of x:")
x.sort()
print(x)

# Demo of argsort which returns the indices of elements (not the elements)
y = np.array([2, 4, 1, 5, 3])
print("Elements of y:")
print(y)

ind = np.argsort(y)
print("Indices of elements:")
print(ind)

# Sorting along rows or cols Demo
x = np.random.randint(0, 10, (4, 6))
print("Two dimensional x:")
print(x)

# Sort elements column-wise
np.sort(x, axis=0)

# Demo - Partial Sorts(Partitioning)
# Sometimes we may want only k smallest elements from an array
# np.partition(theArray, k) gives you k smallest values in left partition and remaining elements in the right partition
x = np.array([9, 7, 2, 4, 5, 8, 1, 3])
print("Given x:")
print(x)
print("Show smallest 3 elements in left partition (not necessarily in any order), leaving other in the right partition:")
print(np.partition(x, 3))

# Demo - Sort along specific rows or cols
x = np.random.randint(0, 10, (4, 6))
print("Two dimensional x:")
print(x)

# sort each column of x
print("the first two slots in each row contain the smallest values from that row, with the remaining values filling the remaining slots:")
print(np.partition(x, 2, axis=1))
```
