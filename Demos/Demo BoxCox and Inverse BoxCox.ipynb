# BoxCox and Reverse BoxCox Demonstration

```python
from scipy import stats
from scipy.special import boxcox, inv_boxcox
import numpy as np

x = np.arange(1, 100)
print('x-values:')
print(x)
print()

x_box, lambda_ = stats.boxcox(x)
print('Lambda = ', lambda_)
print('BoxCoxed x:')
print(x_box)
print()

rev_BoxCox = inv_boxcox(x_box, lambda_)
print('Reverse BoxCoxed Predictions:')
print(rev_BoxCox)
```
