```python
import requests

page = requests.get("https://www.youracclaim.com/users/anandachetan-elikapati/badges")
page.status_code
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
all_courses = soup.find_all(class_="cr-standard-grid-item-content__title")
a_course = all_courses[0]
print(a_course.prettify())
```

```python
titles = [c.get_text() for c in all_courses]
titles
```
