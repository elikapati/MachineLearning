# Demonstration of LaTEX formatting in graph legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = x ** 2
```

```python
# Legend being labelled in LaTEX form of syntax
plt.plot(x, y, label='$y = x^2$')

# Show legend in the best possible space
plt.legend(loc="best", )
```
