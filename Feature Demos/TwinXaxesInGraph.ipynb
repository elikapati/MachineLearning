# Demonstration of Twin x (Having two x - axes in the same graph)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = x ** 2
z = x
```

Plot both the graphs viz., $y = x^2$ and $y = x$ on **same** axes

```python
# Plot both the graphs
plt.plot(x, y)
plt.plot(x, z)
```

Plot both the graphs viz., $y = x^2$ and $y = x$ on **twin y** axes

```python
fig, graph1 = plt.subplots()
graph2 = graph1.twinx()

graph1.plot(x, y, 'g', label = '$y = x^2$')
graph2.plot(x, z, 'b', label = '$y = x$')
fig.legend(loc="center")
```
