```python
import numpy as np
import matplotlib.pyplot as plt

# Returns num evenly spaced samples, calculated over the interval [start, stop]
x = np.linspace(-10, 10, 50)
siny = np.sin(x)
cosy = np.cos(x)
```

```python
# We've not mentioned the size of the figure
figure_obj1 = plt.figure()

# Add plots dividing figure object as 1 row, 2 column and add 1st plot
plt1 = figure_obj1.add_subplot(1, 2, 1)
plt1.plot(x, siny)

# Add plots dividing figure object as 1 row, 2 column and add 2nd plot
plt2 = figure_obj1.add_subplot(1, 2, 2)
plt2.plot(x, cosy)
```

```python
# We've not mentioned the size of the figure
figure_obj2 = plt.figure()

# Add plots dividing figure object as 1 row, 2 column and add 1st plot
plt1 = figure_obj2.add_subplot(2, 1, 1)
plt1.plot(x, siny)

# Add plots dividing figure object as 1 row, 2 column and add 2nd plot
plt2 = figure_obj2.add_subplot(2, 1, 2)
plt2.plot(x, cosy)
```

```python
# first figure
y = np.sin(x)
plt.figure(1)
plt.plot(x,y)
plt.title('Fig1: Sine Curve')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# second figure
plt.figure(2)
y=np.cos(x)
plt.plot(x,y)
plt.title('Fig2: Cosine Curve')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
