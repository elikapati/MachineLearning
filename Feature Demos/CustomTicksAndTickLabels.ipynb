```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (1, 50, 50)
y = x

plt.plot(x, y)
```

```python
# Get Current Axes (=gca)
axes = plt.gca()

# Set x-axis range
axes.xaxis.set_ticks(np.arange(10, 30, 10))
# Set y-axis range
axes.yaxis.set_ticks(np.arange(5, 15, 3))

# Set y-tick labels
axes.yaxis.set_ticklabels(["Five", 8, 'Eleven', 14])

plt.plot(x, y)
```
