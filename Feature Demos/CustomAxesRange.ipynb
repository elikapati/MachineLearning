# Demonstration of Custom X and Y Axis ranges

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x axis range of 50 numbers starting at 1 through 10
x = np.linspace(1, 10, 50)

base_e = np.log(x)
base_2 = np.log2(x)
base_10 = np.log10(x)
```

Plot graphs for $log_e x, log_2 x, log_{10} x$

```python
plt.plot(x, base_e, 'r', x, base_2, 'g', x, base_10, 'b')
```

```python
# Define new ranges for x-axis as [0 - 2] and y-axis as [-1, 2]
plt.axis([0, 5, -1, 5])
plt.plot(x, base_e, 'r', x, base_2, 'g', x, base_10, 'b')
```
