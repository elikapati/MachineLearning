# Demonstration on Annotating

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y = x ** 2

min_x = 0
min_y = min_x ** 2

plt.plot(x, y, min_x, min_y, "ro")

plt.text(0, 60, "Parabola\n$y = x^2$", fontsize=15, ha='center')
plt.text(min_x, min_y+2, "Minima", ha='center')
plt.text(min_x, min_y-4, "(%0.1f, %0.1f)"%(min_x, min_y), ha='center',color='gray')
```
