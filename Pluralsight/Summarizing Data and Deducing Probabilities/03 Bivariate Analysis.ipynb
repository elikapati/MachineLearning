# Bivariate Analysis

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
admission_data = pd.read_csv('Data/Admission_Predict.csv')
admission_data.head(10)
```

```python
# We use "scatter plot" to visualize the relationship between pair of variables
plt.figure(figsize=(10, 8))

plt.scatter(admission_data['GRE Score'], admission_data['Chance of Admit '])

plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')

plt.show()
```

```python
 plt.figure(figsize=(10, 8))

plt.scatter(admission_data['TOEFL Score'], admission_data['Chance of Admit '])

plt.xlabel('TOEFL Score')
plt.ylabel('Chance of Admit')

plt.show()
```

```python
 plt.figure(figsize=(10, 8))

plt.scatter(admission_data['CGPA'], admission_data['Chance of Admit '])

plt.xlabel('CGPA')
plt.ylabel('Chance of Admit')

plt.show()
```

```python
# Let us use the same keeping one of the variables 'Categorical'
plt.figure(figsize=(10, 8))

plt.scatter(admission_data['TOEFL Score'], admission_data['University Rating'])

plt.xlabel('TOEFL Score')
plt.ylabel('University Rating')

plt.show()
```

```python
admission_data_correlation = admission_data.corr()
admission_data_correlation
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(admission_data_correlation, annot=True)
plt.show()
```
