```python
import numpy as np
import pandas as pd
import math

# Creating a 2 dimensional numpy array
data= np.arange(1, 51, 1)

#Creating pandas dataframe from numpy array
df = pd.DataFrame({'A':data})
print("oiginal Data :\n", df)

df_x = np.array(df['A'])
df_x_firstpart = df_x[:-5]
print("df_x_firstpart:\n", df_x_firstpart)

df_x_secondpart = df_x[-5:]
print("df_x_secondpart:\n", df_x_secondpart)

# Creating new column fulled with shifted values
df['B'] = df.shift(math.ceil(len(data)*0.1))
print("Filled new column B with 10% shifted rows of data of column A :\n", df)

# Drop NAN (Not a Number) values
df.dropna(inplace=True)
print("Dataframe after dropping NA values:\n", df)
```
